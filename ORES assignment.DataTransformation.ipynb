```python
political_data.drop(['version','notes'] , axis=1, inplace=True)
political_data.head()
```

```python
political_data['year'] = pd.to_datetime(political_data['year'], format='%Y')
political_data.head()
```

```python
political_data[['candidate', 'party_detailed']] = political_data[['candidate', 'party_detailed']].fillna('UNKNOWN')
political_data.head()
```

```python
political_data['vote_percentage'] = (political_data['candidatevotes'] / political_data['totalvotes']) * 100
political_data.head()
```

```python
political_data.drop(['vote_share'] , axis=1, inplace=True)
political_data.head()
```

```python
political_data['vote_percentage'] = (political_data['candidatevotes'] / political_data['totalvotes']) * 100
political_data.head()
```

```python
political_data.drop_duplicates(inplace=True)
political_data.head()
```

```python
Q1 = political_data.select_dtypes(include=["number"]).quantile(0.25)
Q3 = political_data.select_dtypes(include=["number"]).quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
political_data
```

```python
political_data_no_outliners=political_data[~(( political_data.select_dtypes(include=["number"]).align(Q1-1.5*IQR,axis=1,copy=False)[0]>0)|( political_data.select_dtypes(include=["number"]).align(Q3+1.5*IQR,axis=1,copy=False)[0]<0)).any(axis = 1)]
```

```python
print("original DataFrame:")
print(political_data)
print("\nDataFrame without outliners:")
print(political_data_no_outliners)
```
